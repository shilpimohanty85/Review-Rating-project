# FLIPKART

```python
url = "https://www.flipkart.com/"
driver.get(url)
time.sleep(2)
#clicking on cancel buttopn 
driver.find_element(By.XPATH,"//div[@class='_2QfC02']/button").click()
```

```python
srch_data = ['Phones','smart watches']
```

```python
title = []
review_text = []
ratings = []
```

```python
def scrap():    
        for i in driver.find_elements(By.XPATH,"//div[@class='t-ZTKy']"):
            review_text.append(i.text)
        for i in driver.find_elements(By.XPATH,"//p[@class='_2-N8zT']"):
            title.append(i.text)
        for i in driver.find_elements(By.XPATH,"//p[@class='_2-N8zT']/../div"):
            ratings.append(i.text)
        return
```

```python
urls=[]
```

```python
for i in srch_data:
    #Find search bar
    srchBar = driver.find_element(By.XPATH,"//div[@class='_3OO5Xc']/input")
    srchBar.clear()
    srchBar.send_keys(i)
    #clicking on search button
    driver.find_element(By.XPATH,"//button[@class='L0Z3Pu']").click()
    time.sleep(3)
    page = []
    for i in driver.find_elements(By.XPATH,"//nav[@class='yFHi8N']/a"):
        page.append(i.get_attribute('href'))
    for i in page[0:4]:
        driver.get(i)
        time.sleep(3)
        items = driver.find_elements(By.XPATH,"//a[@class='_1fQZEK']")
        for i in items:
            urls.append(i.get_attribute('href'))
```

```python
len(urls)
```

```python
for i in urls:
    driver.get(i)
    time.sleep(2)
    for _ in range(2):
        driver.execute_script("window.scrollBy(0,6000)")
        time.sleep(1)
        #clicking on all reviews
    try:
        btn=driver.find_element(By.XPATH,"//div[@class='_2c2kV-']/following::a")
        lnk = btn.get_attribute('href')
        driver.get(lnk)
        time.sleep(1)
    except NoSuchElementException:
        pass

    scrap()        
    try:
        n_page=driver.find_elements(By.XPATH,"//nav[@class='yFHi8N']/a")
        np=[]
        for i in n_page:
            np.append(i.get_attribute('href'))
        for i in np[0:18]:
            driver.get(i)
            time.sleep(1)
            scrap()
    except: continue
```

```python
len(ratings), len(review_text), len(title)
```

```python
#creating a dataframe
data = list(zip(title,review_text,ratings))
df2 = pd.DataFrame(data, columns = ["Review_Title","Reiew_Text","Ratings"])
df2
```

```python
#saving the data into csv file
df2.to_csv("flipkrt1_reviews.csv")
```

# AMAZON

```python
# Activating the chrome browser
s=Service("chromedriver.exe")  #opening chrome
driver = webdriver.Chrome(service=s)    
time.sleep(3)
```

```python
url = "https://www.amazon.in/"
driver.get(url)
time.sleep(3)
```

```python
#Searching headphones in the search bar and clicking the search button
search_bar=driver.find_element_by_id('twotabsearchtextbox')
search_bar.send_keys("headphones")

driver.find_element_by_id('nav-search-submit-button').click()
```

```python
#Creating empty lists
Product_URL=[]
Ratings=[]
Review=[]

#Getting URLs of the product
for i in range(1,4):
    URL = driver.find_elements(By.XPATH,"//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']/a")
    for i in URL:
        Product_URL.append(i.get_attribute('href'))
    try:
        next_btn=driver.find_element(By.XPATH,"//li[@class='a-last']/a").click()
    except NoSuchElementException:
        pass
    
for i in Product_URL:
    driver.get(i)
    
    #Clicking the rating
    try:
        driver.find_element(By.XPATH,"//a[1][@id='acrCustomerReviewLink']").click()
    except NoSuchElementException:
        print("No rating")
        pass
    
    #Clicking to see all reviews
    try:
        driver.find_element(By.XPATH,"//a[@class='a-link-emphasis a-text-bold']").click()
    except NoSuchElementException:
        pass
    
    #Scrapping the details
    Start_page=1
    End_page=30
    for page in range(Start_page,End_page+1):
        try:
            Reviews=driver.find_elements(By.XPATH,"//div[@class='a-row a-spacing-small review-data']/span/span")
            for r in Reviews:
                Review.append(r.text.replace('\n',''))
        except NoSuchElementException:
            Review.append("Not Available")
        try:
            Rating=driver.find_elements(By.XPATH,"//div[@class='a-section celwidget']/div[2]/a[1]")
            for i in Rating:
                rating=i.get_attribute('title')
                Ratings.append(rating[:3])
        except NoSuchElementException:
            Ratings.append("Not available")      
            
        #Printing the page scrapped
        #print("Product review and rating  of page {} scraped ".format(page+1))
        
        #Looping for going to next page automatically
        try:
            next_page=driver.find_element(By.XPATH,"//div[@id='cm_cr-pagination_bar']/ul/li[2]/a")
            if next_page.text=='Next Page':
                next_page.click()
                time.sleep(2)
        except NoSuchElementException:
            pass
```

```python
#Checking the length of data extracted
print(len(Review),len(Ratings))
```

```python
#Saving in dataframe
df5=pd.DataFrame({'Review_Text':Review[:12000],'Ratings':Ratings[:12000]})
df5
```

```python
#saving the data into csv file
df5.to_csv("amazon1_reviews.csv")
```
